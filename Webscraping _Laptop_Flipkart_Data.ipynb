**Scraping the webpage**

```python
title = []
rating = []
price = []
features = []


for i in range(1,66):
  URL = 'https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off&page={}'.format(i)

  all_pages = requests.get(URL)
  htmlCode = all_pages.text

  soup = BeautifulSoup(htmlCode, 'html.parser')


  for x in soup.find_all('div', attrs = {'class' : '_2kHMtA'}):
    rate = x.find('div', attrs = {'class' : '_3LWZlK'})
    if rate is None:
      rating.append(np.NaN)
    else:
      rating.append(rate.text)


    product = x.find('div', attrs={'class' : '_4rR01T'})
    if product is None:
      title.append(np.NaN)
    else:
      title.append(product.text)


    MRP = x.find('div', attrs={'class' : '_30jeq3 _1_WHN1'})
    if MRP is None:
       price.append(np.NaN)
    else:
       price.append(MRP.text)

    f = x.find('ul', attrs={'class' : '_1xgFaf'})
    if f is None:
        features.append(np.NaN)
    else:
        features.append(f.text)


print(len(title))
print(len(rating))
print(len(features))
print(len(price))
```

```python
#saving data into csv file

df = pd.DataFrame({'Product': title, 'Rating' : rating, 'Price' : price, 'Feature' : features})
print(df.shape)
print(df.info())
```
